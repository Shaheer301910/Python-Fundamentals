Operators are used to perform operations on variables and values.

Python divides the operators in the following groups:
- `Arithmetic` operators
- `Assignment` operators
- `Comparison` operators
- `Logical` operators
- `Identity` operators
- `Membership` operators
- `Bitwise` operators


__`Arithmetic Operators`__

Arithmetic operators are used with numeric values to perform common mathematical operations.



```python
a = 5
b = 10 

print('Addition: ' , a + b)              
print('Subtraction: ' , b - a)           
print('Multiplication: ' , a * b)
print('Division: ' , b / a)
print('Modulus: ' , b % a)
print('Exponentiation: ' , a ** 2)

c = 143
d = 6

print('Simple divion: ', c / d)
print('Floor division: ' , c // d)
```

The `//` operator computes floor division in which the remainder is discarded

The `%` operator computes the remainder of a floor division.

__`Assignment Operators`__

Assignment operators are used to assign values to variables.

__`Comparison Operators`__

Comparison operators are used to compare two values.

__`Logical Operators`__

Logical operators are used to combine conditional statements.

__`Identity Operators`__

Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.

__`Membership Operators`__

Membership operators are used to test if a sequence is presented in an object.

__`Bitwise Operators`__

Bitwise operators are used to compare (binary) numbers.
